```python
import pandas as pd
import matplotlib.pyplot as plt

fish_data_df = pd.read_csv('/Fish/FIsh-data-anaylisis/Fish2.csv',
                               header=None,
                               names=['Species', 'Weight', 'Length', 'Diagonal', 'Width'])
def showCharts():
    fish_data_df.plot(
        kind='scatter',
        x='Species',
        y='Weight',
        color='blue',
        alpha=0.5,
        title= 'weight of fish by species'
    )
    plt.show()

showCharts()
```
