Data Dictionaries
A data dictionary is a centralized repository of information about data. It provides a detailed description of the data elements used in a system or database, including their meaning, relationships, format, and usage. In essence, it serves as a reference guide that helps users and developers understand the structure and content of the data. 
Each row in a data dictionary represents a field (or column) in a dataframe / database/
